Целое число в базе $a$ будет делится на $(n-1)$ или на множители $(n-1)$ только тогда, когда сумма цифр чисел в базе $a$ будет тоже делиться на $(n-1)$ или на множители $(n-1)$.

$$
\sum\limits_{k=0}^{m}{n^{k}\cdot a_{k}}=\sum_{k=0}^{m}{a_{k}}+(n-1)\cdot\left[\sum_{p=1}^{m}\sum_{k=p}^{m}{n^{k-p}\cdot a_{k}}\right]
$$

Целое число в базе $a$ будет делится на $(n+1)$ или на множители $(n+1)$ только тогда, когда сумма цифр чисел **с переменным знаком** в базе $a$ будет тоже делиться на $(n+1)$ или на множители $(n+1)$.
