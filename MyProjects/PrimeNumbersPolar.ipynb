# Описание 

Программа генерирует большое количество простых чисел и рисует их на поверхности точками, координаты которых заданы так:
$$\Large
\begin{array}{l}
x=\cos{(p)}\cdot p\\
y=\sin{(p)}\cdot p\\
\end{array},\ p\text{ - простое число}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
from math import sin, cos
#------- prime
def Prime(a):
    flag=True
    for i in range(2,a):
        if a%i==0 and i!=a:
            flag=False
            break
    return flag

z = []

for i in range(2,40000):
    if Prime(i):
        z.append(i)

print(len(z))
x = [i*sin(i) for i in z]
y = [i*cos(i) for i in z]


plt.plot(x, y, 'ko', markersize=0.5)
plt.show()
```
