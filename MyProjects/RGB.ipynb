```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
```

```python
def drawGraph(ax, a,b):
    X = np.linspace(0, 200, 400)
    A = (((np.sin(a*X)+1)/2))
    R = ((np.sin(b*X)+1)/2)*(((np.sin(a*X)+1)/2))
    G = ((np.sin(b*X+(2*np.pi/3))+1)/2)*(((np.sin(a*X)+1)/2))
    B = ((np.sin(b*X+(4*np.pi/3))+1)/2)*(((np.sin(a*X)+1)/2))

    # plt.clf()

    ax.plot(X, R+2, c="red")
    ax.plot(X, G+3, c="green")
    ax.plot(X, B+4, c="blue")

    ax.plot(X, R, c="red")
    ax.plot(X, G, c="green")
    ax.plot(X, B, c="blue")
    ax.plot(X, A, '--', c="black")

    # plt.show()

# drawGraph(0.8, 0.4)
```

```python

def update(val):
    ax.clear()
    diffX = diffX_slider.val
    diffY = diffY_slider.val    
    drawGraph(ax, diffX, diffY)
    plt.draw()

fig, ax = plt.subplots(figsize=(12, 8))
plt.subplots_adjust(bottom=0.3)

#   
color = "blue"

diffX = 0.8
diffY = 0.4

drawGraph(ax, diffX, diffY)
slider_size = (0.2, 0.7)

#  
diffX_slider = Slider(
    plt.axes([slider_size[0], 0.15, slider_size[1], 0.03]), 
    'Difference X', 0, 4, 
    valinit=diffX)
diffY_slider = Slider(
    plt.axes([slider_size[0], 0.1, slider_size[1], 0.03]), 
    'Difference Y', 0, 4, 
    valinit=diffY)

diffX_slider.on_changed(update)
diffY_slider.on_changed(update)

plt.show()
```
