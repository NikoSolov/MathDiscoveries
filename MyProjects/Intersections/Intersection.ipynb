Представим на плоскости две прямые уравнениями:
$$
\begin{array}{l}
U(x, t)=\tan(t)(x-M_{1x})+M_{1y}\\
V(x, t)=\tan(\beta\cdot t+\phi)(x-M_{2x})+M_{2y}\\
\end{array}
$$
где $M_{1}, M_{2}$ - центры вращения

Тогда уравнение точки пересечения будет:
$$
D\left(t\right)=\left(
\begin{array}{c}
\frac{\left(A\left(t\right)\cdot M_{1x}-B\left(t\right)\cdot M_{2x}+M_{2y}-M_{1y}\right)}{\left(A\left(t\right)-B\left(t\right)\right)}\\

\end{array}
\right)
$$
где $A(t)=\tan{t},\; B(t)=\tan(\beta\cdot t+\phi)$


# Гипотеза 1
При сонаправленном вращении с одинаковой скоростью прямых, с разностью в угле вращения $\phi,\;\phi\ne0$, точка пересечения за весь период проводит на плоскости окружность
# Задача 1
Найти радиус этой окружности
## Решение задачи 1
Пусть $M_1$ и $M_2$ точки вращения этих прямых. Тогда при вращении угол между этими прямыми состаляет $\phi$ или $\pi-\phi$. Если рассматривать треугольник $M_1M_2I$, где $I$ - точка пересечения прямых, то $\angle M_1IM_2=\phi$ лежит против прямой $M_1M_2$. По теореме синусов можно узнать радиус описанной окружности:
$$
    \dfrac{\|M_1-M_2\|}{\sin{\phi}}=2R \Rightarrow \boxed{R = \dfrac{\|M_1-M_2\|}{2\sin{\phi}}}
$$
> Заметим, что $\sin(\phi)=\sin(\pi-\phi)$

# Задача 2
Найти координаты центра этой окружности и записать её формулу

## Решение частного случая задачи 2 $\left(\phi = \dfrac{\pi}{2}\right)$
Прямые лежат диаметрально противополжно. В таком случае координаты центра будут середины между точками каждого компонента координаты
$$
C=\left(
    \dfrac{M_2.x+M_1.x}{2},
    \dfrac{M_2.y+M_1.y}{2}
\right)
$$
Тогда уравнение окружности примет вид:
$$
    \left(x-\dfrac{M_2.x+M_1.x}{2}\right)^2+\left(y-\dfrac{M_2.y+M_1.y}{2}\right)^2=\dfrac{\|M_1-M_2\|^2}{4}
$$

## Ответ к Задаче 2
$$
\left(x-\frac{M_{2}.x+M_{1}.x}{2}+\frac{\cot\left(\phi\right)\left(M_{2}.y-M_{1}.y\right)}{2}\right)^{2}+\left(y-\frac{M_{2}.y+M_{1}.y}{2}-\frac{\cot\left(\phi\right)\left(M_{2}.x-M_{1}.x\right)}{2}\right)^{2}=\left(\frac{\|M_1-M_2\|}{2\sin\left(\phi\right)}\right)^{2}
$$

```python
from math import radians, sin, cos,tan
import pygame as pg
from random import randint as rnd
from random import choice
#------------------------
width,height=512,512
lineWidth = 3
dotSize = 3
alpha = 0
lineCount = 3
#------------------------

root=pg.display.set_mode((width, height))
clock=pg.time.Clock()
game=True

class vec2():
    a=[]
    def __init__(self,x,y,angle,angleSpeed,color):
        self.x,self.y = x,y   # центр вращения прямой
        self.angle = angle    # угол прямой
        self.angleSpeed = angleSpeed # угловая скорость
        self.color = pg.Color(color) # цвет
        vec2.a.append(self)          
    def draw(root):
        for i in vec2.a:
            if (i.angle==90):
             pg.draw.line(root, i.color, (i.x, 0), (i.x, height),2)
            elif (i.angle == 0):
             pg.draw.line(root, i.color, (0, i.y), (width, i.y),2)
            else:
             k=tan(radians(i.angle))

             xUp=-(i.y)/k+i.x
             xDown=(height-i.y)/k+i.x
             pg.draw.line(root, i.color, (xUp, 0), (xDown, height),lineWidth)

#             yLeft=k*(-i.x)+i.y
#             yRight=k*(width-i.x)+i.y
#             pg.draw.line(root, i.color, (0, yLeft), (width, yRight),lineWidth)

             pg.draw.circle(root, i.color, (i.x,i.y), dotSize)
    #def clear():
       
    def update():
     for i in vec2.a:
      i.angle+=i.angleSpeed


    def draw_int(root):
        for b in range(len(vec2.a)):
            for c in range(b,len(vec2.a)):
                k1 = tan(radians(vec2.a[b].angle))
                k2 = tan(radians(vec2.a[c].angle))                
                if (k1-k2!=0):
                   
                   Xi = (k1*vec2.a[b].x - k2*vec2.a[c].x + (vec2.a[c].y-vec2.a[b].y))/(k1-k2)
                   Yi = k2*(Xi-vec2.a[c].x)+vec2.a[c].y
                   if ((Xi<=width and Xi>=0) and (Yi<=height and Yi>=0)):
                     pg.draw.circle(root, (vec2.a[b].color+vec2.a[c].color), (Xi,Yi), dotSize)

fade=pg.Surface((width, height), pg.SRCALPHA)
surf=pg.Surface((width, height))
fade.fill((0,0,0))
fade.set_alpha(alpha)

for i in range(lineCount):
    vec2(rnd(width//3,width//3*2),rnd(height//3,height//3*2),rnd(0,360),choice([1,-1,2,-2,3,-3]), (rnd(0,255),rnd(0,255),rnd(0,255)))

flag=True

while game:
    for event in pg.event.get():
        if event.type==pg.QUIT or (event.type==pg.KEYDOWN and event.key==pg.K_ESCAPE):
            game=False
        if event.type==pg.KEYDOWN and event.key==pg.K_SPACE:
            flag=not(flag)
        if event.type == pg.KEYDOWN and event.key == pg.K_RETURN:
            surf.fill((0,0,0))
            vec2.a=[]
            for i in range(lineCount):
                vec2(rnd(width//3,width//3*2),rnd(height//3,height//3*2),rnd(0,360),choice([1,-1,2,-2,3,-3]), (rnd(0,255),rnd(0,255),rnd(0,255)))

            
    root.fill((0,0,0))
    surf.blit(fade,(0,0))

    vec2.draw_int(surf)
    root.blit(surf,(0,0))
    vec2.draw(root)
    if (flag): vec2.update()
    pg.display.update()
    clock.tick(60)

pg.quit()
```
