$$
\text{Leg}_{n}(m\cdot n^{a})=a
$$

```python
def leg(m, n):
    p = m
    a = 0
    while (p%n==0):
        p//=n
        a+=1
    return a, p
leg(6,2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(16,4))
for j in range(1,4):
    d = 2**j
    x = np.arange(1, 100)
    y = [leg(i, d)[0] for i in x]
    plt.scatter(x,y, s=10)
    plt.plot(x,y, '--')
```
