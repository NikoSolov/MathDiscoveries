# Описание 

Программа изображает графику как из этого [видео](https://www.youtube.com/watch?v=-qgreAUpPwM&pp=ygUbM2JsdWUxYnJvd24gZm91cmllciBzZXJpZXMg).

# Формулы

$$\Large
\begin{array}{l}
X(t)=\sum\limits_{n=1}^{k}{(A_{X}[k]\cdot\cos(F_{X}[k]\cdot t+O_{X}[k]))}\\
Y(t)=\sum\limits_{n=1}^{k}{(A_{Y}[k]\cdot\sin(F_{Y}[k]\cdot t+O_{Y}[k]))}\\
A_{X}[\ ], A_{Y}[\ ] - \text{амплитуды (длины плечей)}\\
F_{X}[\ ], F_{Y}[\ ] - \text{частоты (скорости вращения плечей)}\\
O_{X}[\ ], O_{Н}[\ ] - \text{фазы (начальные углы плечей)}
\end{array}
$$
# Модель Desmos


```python
import pygame as pg
from random import randint as rd
from random import choice as ch
from math import cos, sin, radians
#----------------------
width, height=1024, 768
minLen, maxLen = 50,100
maxDAngle = 20
lineCount = 4
lineThickness = 3
alpha = 0
thin = 10
#----------------------
root=pg.display.set_mode((width, height))
clock=pg.time.Clock()
game=True

class vector():
 startPos=(width//2, height//2) # позиция корня плечей
 lastPos=[0,0]              # последняя позиция точки
 arr=[]
 def __init__(self, length, angle, d_angle,color):
  self.length=length    # длина одного плеча
  self.angle=angle      # угол
  self.d_angle=d_angle  # изменение 
  self.color=color      # цвет плеча
  vector.arr.append(self)
  # ------- получение начальной позиции точки ----------------
  x,y = vector.startPos[0], vector.startPos[1]
  for i in vector.arr:
   x+=i.length*cos(radians(i.angle))
   y+=i.length*sin(radians(i.angle))
  vector.lastPos=[x,y]

 def draw_vecs(root):
  x,y = vector.startPos[0], vector.startPos[1]
  for i in vector.arr:
   pg.draw.line(root, i.color, (x,y), 
                               (x+i.length*cos(radians(i.angle)),
                                y+i.length*sin(radians(i.angle))),lineThickness)
   x+=i.length*cos(radians(i.angle))
   y+=i.length*sin(radians(i.angle))

 def update(k):
  vector.last_x, vector.last_y = vector.startPos[0], vector.startPos[1]
  for i in vector.arr:
   i.angle+=i.d_angle*k

 def draw(root, color):
  col=pg.Color(0,0,0)
  col.hsva=color
  x,y=vector.startPos
  for i in vector.arr:
   x+=i.length*cos(radians(i.angle))
   y+=i.length*sin(radians(i.angle))
  pg.draw.circle(root, col, (int(x),int(y)), thin//2)
  vector.lastPos=[int(x),int(y)]

 def draw_line(root, color):
  col=pg.Color(0,0,0)
  col.hsva=color
  x,y=vector.startPos
  for i in vector.arr:
   x+=i.length*cos(radians(i.angle))
   y+=i.length*sin(radians(i.angle))
  pg.draw.line(root, col, vector.lastPos, (int(x),int(y)), thin)
  vector.lastPos=[int(x),int(y)]

   
for i in range(lineCount):
     vector(rd(minLen,maxLen), rd(0,359), ch([rd(0,maxDAngle), (-1)*rd(0,maxDAngle)]),(rd(0,255),rd(0,255),rd(0,255)))

surf=pg.Surface((width,height))
fade=pg.Surface((width,height), pg.SRCALPHA)
surf.fill((0,0,0))
fade.fill((0,0,0))
fade.set_alpha(alpha)
k=0.1
i=0
while game:
 for event in pg.event.get():
  if event.type==pg.KEYDOWN:
   if event.key==pg.K_ESCAPE:
    game=False
   if event.key==pg.K_UP: k+=0.1
   if event.key==pg.K_DOWN: k-=0.1
   if event.key==pg.K_SPACE:
    vector.arr=[]
    for i in range(rd(2,lineCount)):
     vector(rd(minLen,maxLen), rd(0,359), ch([rd(0,maxDAngle), (-1)*rd(0,maxDAngle)]),(rd(0,255),rd(0,255),rd(0,255)))
    surf.fill((0,0,0))

# root.fill((0,0,0))
 surf.blit(fade,(0,0))
 vector.update(k)
 vector.draw_line(surf,[i%361,100,50,0])
 vector.draw(surf,[i%361,100,50,0])
 root.blit(surf, (0,0))
 vector.draw_vecs(root)
 
 pg.display.flip()
 pg.display.set_caption(str(clock.tick(60)))
 i+=1
pg.quit()
```
