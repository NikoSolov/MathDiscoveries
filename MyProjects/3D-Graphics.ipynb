# График
Рисует график $\Large y=5\cdot\sin(x\cdot\frac{\pi}{18})+\frac{(z-5)^2}{2}+10$

```python
import pygame as pg
from math import cos, sin, radians, degrees
from random import randint as rd
win={"width":1280, "height":768}
root=pg.display.set_mode(list(win.values()))
clock=pg.time.Clock()
game=True


def draw_ray(begin,step, size, angle,color):
 i=0
 while ((0<int(begin[0]+cos(radians(angle))*size//2+sin(radians(angle))*step*(i))<win["width"]) and
        (0<int(begin[1]+sin(radians(angle))*size//2-cos(radians(angle))*step*(i))<win["height"]) and
        step!=0):
 # if step!=0:
  pg.draw.line(root, color,
              (int(begin[0]+cos(radians(angle))*size//2+sin(radians(angle))*step*i),
               int(begin[1]+sin(radians(angle))*size//2-cos(radians(angle))*step*i)),
              (int(begin[0]-cos(radians(angle))*size//2+sin(radians(angle))*step*i),
               int(begin[1]-sin(radians(angle))*size//2-cos(radians(angle))*step*i)))
  i+=1 
 pg.draw.line(root,color, begin,
              (int(begin[0]+sin(radians(angle))*step*i),
               int(begin[1]-cos(radians(angle))*step*i)))

 return i
 
i=0
angle=0
xstep=20
ystep=40
zstep=10
xangle=90
yangle=0
zangle=225
center=[win["width"]//2,  win["height"]//2]

a=10
b=10

while game:
 #try:
 for event in pg.event.get():
  if event.type==pg.KEYDOWN:
   if event.key==pg.K_ESCAPE:
    game=False

 key=pg.key.get_pressed()
 if key[pg.K_UP]: center[1]-=30
 if key[pg.K_DOWN]: center[1]+=30
 if key[pg.K_LEFT]: center[0]-=30
 if key[pg.K_RIGHT]: center[0]+=30
 if key[pg.K_w]: ystep+=1
 if key[pg.K_s]: ystep-=1
 if key[pg.K_a]: xstep-=1
 if key[pg.K_d]: xstep+=1
 if key[pg.K_q]: xangle-=1
 if key[pg.K_e]: xangle+=1
 if key[pg.K_z]: yangle-=1
 if key[pg.K_x]: yangle+=1
 if key[pg.K_c]: zstep-=1
 if key[pg.K_v]: zstep+=1
 if key[pg.K_f]: zangle-=1
 if key[pg.K_g]: zangle+=1

 root.fill((0,0,0))
 pg.draw.circle(root, (255,255,255), center, 5)
 max_y=draw_ray(center, ystep, 20,yangle,(255,0,0))
 max_x=draw_ray(center, xstep,20, xangle, (0,255,0))
 max_z=draw_ray(center, zstep,20, zangle, (0,0,255))
 

 for z in range(15):
  color=[0,105+z*10, 105+z*10]
  for x in range(300):
   y=sin(radians(x*10))*5+10+(z-5)**2//2
   zd=z
   pg.draw.circle(root, color,
                  (int(center[0]+x*xstep*cos(radians(yangle))-y*ystep*cos(radians(xangle))+zd*zstep*sin(radians(zangle))),
                   int(center[1]+x*xstep*sin(radians(yangle))-y*ystep*sin(radians(xangle))-zd*zstep*cos(radians(zangle)))
                   ), 5)

 pg.display.flip()
 clock.tick(30)
print(max_y, max_x) 
print("done")
pg.quit()
```
