## Правила дифференцирования
### Правило линейности
$$
\begin{array}{l}
[a\cdot f(x)+b]'=\\=\lim\limits_{\Delta x\to0}{\dfrac{a\cdot f(x+\Delta x)\cancel{+b}-a\cdot f(x)\cancel{-b}}{\Delta x}}=\\
=a\cdot \lim\limits_{\Delta x\to0}{\dfrac{\Delta f(x)}{\Delta x}}=a\cdot f'(x)
\end{array}
$$

### Производная произведения
$$\small
\begin{array}{l}
[f(x)\cdot g(x)]'=\\
\lim\limits_{\Delta x\to0}{\dfrac{f(x+\Delta x)\cdot g(x+\Delta x)-f(x)\cdot g(x)}{\Delta x}}\\
\text{Так как }f(x+\Delta x)-f(x)=\Delta f(x)\\
\text{значит }f(x+\Delta x)=f(x)+\Delta f(x)\\ 
\lim\limits_{\Delta x\to0}{\dfrac{f(x+\Delta x)\cdot g(x+\Delta x)-f(x)\cdot g(x)}{\Delta x}}=\\
=\lim\limits_{\Delta x\to0}{\dfrac{(f(x)+\Delta f(x))\cdot (g(x)+\Delta g(x))-f(x)\cdot g(x)}{\Delta x}}\\
=\lim\limits_{\Delta x\to0}{\dfrac{\cancel{(f(x)\cdot g(x)}+\Delta f(x)\cdot g(x)+\Delta g(x)\cdot f(x)+\Delta f(x)\cdot \Delta g(x))\cancel{-(f(x)\cdot g(x)}}{\Delta x}}\\
=g(x)\cdot \lim\limits_{\Delta x\to 0}{\dfrac{\Delta f(x)}{\Delta x}}+
f(x)\cdot \lim\limits_{\Delta x\to 0}{\dfrac{\Delta g(x)}{\Delta x}}+
\lim\limits_{\Delta x\to 0}{\dfrac{\Delta f(x)\Delta g(x)}{\Delta x}}=\\
=g(x)\cdot \lim\limits_{\Delta x\to 0}{\dfrac{\Delta f(x)}{\Delta x}}+
f(x)\cdot \lim\limits_{\Delta x\to 0}{\dfrac{\Delta g(x)}{\Delta x}}+
\lim\limits_{\Delta x\to 0}{\dfrac{\Delta f(x)}{\Delta x}}\cdot \lim\limits_{\Delta x\to0}{\Delta g(x)}=\\
=g(x)\cdot f'(x)+ f(x)\cdot g'(x)+ f'(x)\cdot 0=\\
=\boxed{g(x)\cdot f'(x)+ f(x)\cdot g'(x)}

\end{array}
$$

### $[f^{n}(x)]'=n\cdot f^{n-1}(x)\cdot f'(x)$
$$
\boxed{
\begin{array}{l}
\lim\limits_{\Delta x\to 0}{\dfrac{(x+\Delta x)^{n}-x^{n}}{\Delta x}}=\\
=\lim\limits_{\Delta x\to 0}{\dfrac{\sum\limits_{k=0}^{n}{\left(\binom{k}{n}x^{n-k}\cdot\Delta x^{k}\right)}-x^{n}}{\Delta x}}=\\
=\lim\limits_{\Delta x\to 0}{\dfrac{\cancel{x^{n}}+\binom{1}{n}x^{n-1}\Delta x+\dots+\binom{n}{n}\Delta x^{n}-\cancel{x^{n}}}{\Delta x}}=\\
=\lim\limits_{\Delta x\to 0}{n\cdot x^{n-1}+\dots+\binom{n}{n}\Delta x^{n-1}}=\boxed{n\cdot x^{n-1}}
\end{array}}
$$

$$
\begin{array}{l}

\boxed{\ln{x}=\lim\limits_{\Delta x\to 0}{\dfrac{x^{\Delta x}-1}{\Delta x}}}\\
\lim\limits_{\Delta x\to0}\dfrac{a^{\Delta x}-1}{\Delta x}=1\to a=\lim\limits_{\Delta x\to0}(\Delta x+1)^\dfrac{1}{\Delta x}\\

\end{array}
$$


#### $[e^{f(x)}]=e^{f(x)}\cdot f'(x)$
$$
\begin{array}{l}

e^{x}\cdot\lim\limits_{\Delta x\to0}{\dfrac{e^{\Delta x}-1}{\Delta x}}=e^{x}\cdot \ln{e}=\boxed{e^{x}}\\
\end{array}
$$
#### $[\log_{a}{f(x)}]'=\dfrac{1}{\ln{a}\cdot f(x)}\cdot f'(x)$
$$
\begin{array}{l}
[\log_{a}{x}]'=\lim\limits_{\Delta x\to 0}{\dfrac{\log_{a}{(x+\Delta x)}-\log_a{(x)}}{\Delta x}}=\\
\lim\limits_{\Delta x\to 0}{\dfrac{\log_{a}{\left(\dfrac{x+\Delta x}{x}\right)}}{\Delta x}}=
\lim\limits_{\Delta x\to 0}{\dfrac{\log_{a}{\left(1+\dfrac{\Delta x}{x}\right)}}{\Delta x}}=\\
\lim\limits_{\Delta x\to 0}{\dfrac{\log_{a}{\left(1+\dfrac{\Delta x}{x}\right)}}{\Delta x}}=
\lim\limits_{\Delta x\to 0}{\dfrac{\dfrac{\Delta x}{x\cdot\ln(a)}}{\Delta x}}=\boxed{\dfrac{1}{x\cdot\ln(a)}}
\end{array}
$$
#### $[\ln{f(x)}]'=\dfrac{1}{f(x)}\cdot f'(x)$
$$
\begin{array}{l}
[\ln{x}]'=\lim\limits_{\Delta x\to0}{\dfrac{\ln(x+\Delta x)-\ln(x)}{\Delta x}}=\\
\lim\limits_{\Delta x\to0}{\dfrac{\ln(x+\Delta x)-\ln(x)}{\Delta x}}=
\lim\limits_{\Delta x\to 0}{\dfrac{\ln{\left(1+\dfrac{\Delta x}{x}\right)}}{\Delta x}}=\\
\lim\limits_{\Delta x\to 0}{\dfrac{\dfrac{\Delta x}{x\cdot\ln(e)}}{\Delta x}}=\boxed{\dfrac{1}{x}}
\end{array}
$$
